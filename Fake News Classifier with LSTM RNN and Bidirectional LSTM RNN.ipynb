### Importing the required libraries

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM, Bidirectional
from tensorflow.keras.layers import Dense
```

### Initializing the vocabulary size

```python
voc_size = 5000
```

```python
message = X.copy()
```

```python
message['title'][1]
```

```python
message
```

```python
message.reset_index(inplace = True)
```

```python
message
```

## Text Preprocessing part

```python
import re
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0,len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus[1]
```

## After the preprocessing we apply ONE HOT ENCODER

```python
onehot_rep = [one_hot(word, voc_size) for word in corpus]
onehot_rep
```

```python
onehot_rep[1]
```

## Padding Representation

```python
sent_length = 20
```

```python
embedded_docs = pad_sequences(onehot_rep, padding = "pre", maxlen = sent_length)
print(embedded_docs)
```

# Creating the model [embedding(for semantic meaning) and LSTM]

```python
## Creating model
embedding_vector_features=40 ##features representation

#embedding
model=Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=embedding_vector_features, input_length=sent_length))

#LSTM model
model.add(LSTM(100))

#now the output is ready for a ANN input
model.add(Dense(1,activation='sigmoid'))

#Manually building the model
model.build(input_shape=(None, sent_length))

#compiling
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.3, random_state=42)
```

# Model Training

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs = 10, batch_size = 100)
```

## Performance Metrics and Accuracy

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred >0.5, 1,0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## Adding droput for improving the accuracy(Reduces overfitting)

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.build(input_shape=(None, sent_length))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs = 10, batch_size = 100)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred >0.5, 1,0)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# BIDIRECTIONAL LSTM RNN

## Applying bidirectional lstm rnn to the above data

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(Bidirectional(LSTM(100)))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.build(input_shape=(None, sent_length))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs = 10, batch_size = 100)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred >0.5, 1,0)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
